## Import libraries

```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read CSV file and sort it

```python
#Remember to change the file name down below as appropriate!
#This should be the exact same file obtained from the hdfs merge command and stored on the local fs.
df = pd.read_csv('out.csv')
df.columns=['SIM','DAY', 'TOTAL']
df['DAY'] =pd.to_datetime(df.DAY)
df = df.sort_values(by=['DAY'], ascending=True)
df.tail()
```

## Plot Monte Carlo simulations graph

```python
sns.set(rc={'figure.figsize':(38,16)})
sns.lineplot(x="DAY", y="TOTAL", hue="SIM", data=df)
```
